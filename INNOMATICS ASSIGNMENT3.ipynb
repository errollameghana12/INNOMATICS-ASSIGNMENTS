```python
#Problem1:EMPLOYEE PERFORMANCE BONUS ELIGIBLITY
# dictionary containing employee names and their performance scores
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
#finding the highest performance score
highest_score=max(employees.values())
#creating a list to store employees with highest score
top_performers=[]
#checking a list to store employees who have highest score
for name,score in employees.items():
    if score == highest_score:
        top_performers.append(name)
#display the result
print("Top Performers Eligible for Bonus:",  ", ".join(top_performers), f"(Score: {highest_score})")
```

```python
#Problem2:SEARCH QUERY KEYWORD ANALYSIS
#input search query
query="Buy mobile phone buy phone online"
#step1: Convert to lowercase
query=query.lower()
#step2:Remove punctuation(if any)
import string 
for char in string.punctuation:
    query=query.replace(char, "")
#step3: split sentence into words
words=query.split()
#step4: count frequency of each word
frequency={}
for word in words:
    if word in frequency:
        frequency[word]+=1
    else:
        frequency[word]=1
#step5: display words searched more than once
result={}
for word,count in frequency.items():
    if count>1:
        result[word]=count
print(result)
```

```python
#Problem3: SENSOR DATA VALIDATION
#list of sensor readings
sensor_readings=[3,4,7,8,10,12,5]
#creating an empty list to store valid readings
valid_readings=[]
#checking each reading using index and value
for index,value in enumerate(sensor_readings):
    #if reading is even,it is valid
    if value%2==0:
        valid_readings.append((index,value))
#displaying the result
print("Valid Sensor Readings (Hour,Value):")
print(valid_readings)
```

```python
#Problem4: Email Domain Usage Analysis
#list of email IDs
emails = ["ravi@gmail.com","anita@yahoo.com","kiran@gmail.com","suresh@gmail.com","meena@yahoo.com"
]
#dictioanary  to store domain count
domain_count={}
#extract domain count frequency
for email in emails:
    domain=email.split("@")[1] #extract domain part
    if domain in domain_count:
        domain_count[domain]+=1
    else:
        domain_count[domain]=1
#calculate percentage usage
total_emails=len(emails)
for domain,count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {int(percentage)}%")
 
```

```python
#Problem6:Duplicate User ID Detection
# List of user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
# Dictionary to store frequency of each user ID
frequency = {}
#Count occurrences of each user ID
for user in user_ids:
    if user in frequency:
        frequency[user] += 1
    else:
        frequency[user] = 1
#Display only duplicate user IDs
for user, count in frequency.items():
    if count > 1:
        print(f"{user} → {count} times")
```
