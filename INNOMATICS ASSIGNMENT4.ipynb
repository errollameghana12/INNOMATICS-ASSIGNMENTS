```python
#prblem1:recharge validation system:
def recharge_validation(recharge_amount,valid_plans):
    plans=[199,299,399,599]
    if recharge_amount<50:
        return "Invalid amount:amount must be >=50"
    if valid_plans not in plans:
        return "Invalid plan:"
    return "Recharge Successful"
while True:
    recharge_amount=int(input("Enter amount:"))
    valid_plans=int(input("Enter your plan:"))
    result=recharge_validation(recharge_amount,valid_plans)
    print(result)
    if result=="Recharge Successful":
        break
    else:
        print("Please Try Again.")
    
```

```python
#problem2:inventory reorder alert system
# Dictionary of products and their stock quantities
inventory = {"Rice": 20,"Sugar": 10,"Oil": 5,"Wheat": 18}
# Function to check stock levels
def check_inventory(products):
    for product, quantity in products.items():
        if quantity < 15:
            print(product, "→ Reorder Alert")
        else:
            print(product, "→ Stock OK")

# Calling the function
check_inventory(inventory)
```

```python
#problme3:student result processing system
def calculate_result(marks):
    total=0
    for mark in marks:
        total+=mark
    average=total/len(marks)
    if average>=50:
        return f"Average:{average} - pass"
    else:
        return f"Average:{average} - Fail"
student_marks=[50,60,75,84]
result=calculate_result(student_marks)
print(result)
```

```python
#Problem4: cab fare estimator with retry option
def cab_fare_calculation(distance,peak_hour):
    base_rate=50
    per_km=12
    fare=base_rate+(distance*per_km)
      # Add 25% extra during peak hour
    if peak_hour.lower()=="yes":
        fare=fare*1.25
    return fare
# Retry option using while loop
while True:
    distance=int(input("Enter distance in km:"))
    peak_hour=input("Is it peak hour?(yes/no):")
    total_fare=cab_fare_calculation(distance,peak_hour)
    print("Total Fare:",total_fare,"/-")
    retry=input("Do you wat to calculate again?(yes/no):")
    if retry.lower()!="yes":
        print("Thank you for using cab Fare Estimator!")
        break
```

```python
#Problem5:Employee Attendence Eligibilty Checker
def check_attendence(attendence_list):
    total_days=len(attendence_list)
    present_days=0
    for days in attendence_list:
        if days=="P":
            present_days+=1
    percentage=(present_days/total_days)*100
    if percentage>=75:
        return f"Attendence: {percentage}% - Eligible"
    else:
        return f"Attendence:{percentage}% - Not Eligible"
# Example attendance list
attendence=["P","A","P","P","A","A","A","P","P","A","P"]
result=check_attendence(attendence)
print(result)
```

```python
def check_password(password):
    has_digit=False
    has_special=False
    special_chars="@#$"
    # Check each character in password
    for char in password:
        if char.isdigit():
            has_digit=True
        if char in special_chars:
            has_special=True
    #apply rules
    if len(password)<8:
        return "Weak password:min length should be 8"
    elif not has_digit:
        return "Weak password:must contain at least one digit"
    elif not has_special:
        return "Weak password: must contains at least one special character(@#$)"
    else:
        return "Strong password"
#input
user_pswrd=input("Enter your password:")
result=check_password(user_pswrd)
print(result)
```
