```python
#PROBLME1: Unique Words in a Sentence
sentence=input("Enter Sentence: ")
#Takes a sentence as input from the user
s1=sentence.lower()
#Converts the sentence to lowercase so that words like "Python " and "python" are treated same
words=sentence.split()
#Splits the sentence into a list of indivisual words based on spaces
unique_words=set(words)
#Converts the list of words into a set to remove duplicate words
print("Unique words count:",len(unique_words))
#Prints the total number of unique words
print("Unique words:",unique_words)
#Prints the set of unique words
```

```python
#PROBLEM2: Highest Salary from Employee Data
employees={"Ravi": 75000,"Anitha": 68000,"Kiran":7200}
#Dictionary storing employee names as keys and their salaries as values 
highest_salary=0
#Variableto keep track of the highest salary found so far
highest_employee=""
#Variable to store the name of the employee with the highest salary
for name,salary in employees.items():
    #Loop through the dictionary to access each employee name and salary
    if salary>highest_salary:
        #Check if the salary is greater than the stored highest salary
        highest_salary=salary
        #Update highest_salary with the crrent salary if it is greater
        highest_employee=name
        #Update highest_employee with the corresponding employee name
print("Highest Salary:",highest_employee,"-",highest_salary)
#Print the employee who has the salary along with the salary amount
```

```python
#PROBLME3: Find Maximum and Minimum Values
numbers=[45,22,89,10,66]
#List of integers
max_value=numbers[0]
min_value=numbers[0]
#Initialize max and min with the first element of the list
for num in numbers[1:]:
    #Loop thorugh the list starting from the second element
    if num>max_value:
        max_value=num
        #if the current number is less than min_value,update min_value
    if num<min_value:
        min_value=num
        #if current number is less than min_value,update min_value
print("List:",numbers)
#display the original list
print("Max:",max_value)
#display the maximum value
print("Min:",min_value)
#display the minimum value
```

```python
#PROBLEM 4:Count Products Above a Price Threshold
#List of product prices
prices=[450,1200,899,1500,300]
#initialize a counter to track products above 100
count=0
#Loop through each price in the list
for price in prices:
    #check if the price is greater than 1000
    if price>1000:
        count+=1 #increment the counter if condition is true
 #print the final count
print("Products above 1000:",count)
```

```python
#PROBLEM 5: Calculate Attendence Percentage
#list representing attendence record
attendence=["P","P","A","P","P"]
#count total no of days
total_days=len(attendence)
#initialize present counter
present_count=0
#loop through attendence list
for status in attendence:
    #check if the student was present
    if status=="P":
        present_count+=1 #increase present count
    #calculate attendence percentage
attendence_percentage=(present_count/total_days)*100
#print the result
print("Attendence Percentage:",attendence_percentage)
```

```python
#PROBLEM 7:Count Character Frequency
#given string
text="pythonp"
#create an empty dictionary to store character counts
char_count={}
#loop thorugh each character in the string
for ch in text:
    #if character already exists in dictionary,increment its count
    if ch in char_count:
        char_count[ch]+=1
    else:
        #if character is seen for the first time,initialize count s 1
        char_count[ch]=1
#print the character frequency dictionary
print(char_count)
```
