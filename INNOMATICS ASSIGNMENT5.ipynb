```python
#Post Engagement Analyzer
def post_engagement(likes):
    total=0
    for like in likes:
        total+=like
    if total>=100:
        post_status="Viral Post"
    else:
        post_status="Normal Engagement"
    return total,post_status
#user input
likes=[]
days=int(input("Enter number of days:"))
for i in range(days):
    value=int(input(f"Enter likes for day {i+1}:"))
    likes.append(value)
total, post_status=post_engagement(likes)
print("Total Likes:",total)
print("Post Status:",post_status)
    
```

```python
#Rainfall Adequacy Checker
def rainfall_analyzer(rainfall_data):
    total=0
    for rain in rainfall_data:
        total+=rain
    average=int(total/len(rainfall_data))
    if average>=70:
        status="Adequate Rainfall"
    else:
        status="Inadequate Rainfall"
    return average,status
#user input
rainfall_data=[]
days=int(input("Enter number of rainfall days:"))
for i in range(days):
    value=int(input(f"Enter rainfall for day {i+1}:"))
    rainfall_data.append(value)
average,status=rainfall_analyzer(rainfall_data)
print("Average Rainfall:",average)
print("Rainfall Status:",status)
```

```python
# Social Media – Duplicate Account Detection
def detect_duplicates(usernames):
    unique_usernames = set(usernames)
    if len(usernames) != len(unique_usernames):
        return "Yes"
    else:
        return "No"
# User input
usernames = []
n = int(input("Enter number of usernames: "))
for i in range(n):
    name = input(f"Enter username {i+1}: ")
    usernames.append(name)
result = detect_duplicates(usernames)
print("Duplicate Accounts Found:", result)
```

```python
# Agriculture – Premium Crop Price Filter
def filter_premium(prices):
    premium_crops = []
    for price in prices:
        if price > 2000:
            premium_crops.append(price)
    
    return premium_crops
# User input
prices = []
n = int(input("Enter number of crops: "))
for i in range(n):
    value = int(input(f"Enter price for crop {i+1}: "))
    prices.append(value)
premium_list = filter_premium(prices)

print("Premium Crops:", premium_list)
```

```python
#Application Health Checker
def health_checker(error_count):
    if error_count==0:
        status="Healthy"
    elif error_count<=5:
        status="Minor Issues"
    else:
        status="Critical Issues"
    return error_count,status
#user input
error_count=int(input("Enter the Error count:"))
error_count,status=health_checker(error_count)
print("Error Count:",error_count)
print("System status:",status)
```

```python
# Daily Transaction Limit Checker
def transaction_limit_validator(transact_amount):
    daily_amount=50000
    if transact_amount<=daily_amount:
        status="Approved"
    else:
        status="Rejected"
    return transact_amount,status
transact_amount=int(input("Enter the amount for transaction:"))
transact_amount,status=transaction_limit_validator(transact_amount)
print("Transaction Amount:",transact_amount)
print("Transaction Status:",status)
```

```python
# Student Attendance Eligibility System

def attendance_checker(attendance_list):
    present_count = 0
    for day in attendance_list:
        present_count += day
        percentage = (present_count / len(attendance_list)) * 100
    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"
    return percentage, eligibility
# user input
attendance = []
days = int(input("Enter total number of Classes: "))
for i in range(days):
    value = int(input(f"Enter attendance for class {i+1} (1=Present, 0=Absent): "))
    attendance.append(value)

percentage, eligibility = attendance_checker(attendance)

print("Attendance Percentage:", percentage)
print("Exam Eligibility:", eligibility)
```

```python
# Smart Electricity Billing System
def calculate_bill(units):
    bill=0
    # Slab calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
  # Usage classification
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status
# User input
units = int(input("Enter total units consumed: "))
total_bill, usage_status = calculate_bill(units)
print("Total Bill: ₹", total_bill)
print("Usage Level:", usage_status)
```
