```python
marks = [45, 78, 90, 33, 60]
p_count=0 # to count pass students
f_count=0 # to count fail students
for i in marks:
    if(i>=50):
        p_count+=1
    else:
        f_count+=1
print(p_count)
print(f_count)        
```

```python
names = [" Alice ", "bob", " CHARLIE "]
cleaned_list=[]
for i in names:
    i=i.strip() #removes extra spaces
    i=i.lower() #converts into lowercase
    cleaned_list.append(i)  #another list to store first list after removing extra space and converting into lowercase
print(cleaned_list)
```

```python
messages = ["Hi", "Welcome to the platform", "OK"]
for i in messages:
    length=len(i)
    print(i,length)
    if(length>10):
        print(i,"Flagged")
```

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
e_count=0 # to count no of ERRORS
for i in logs:
        if(i=="ERROR"):
            e_count+=1
print(e_count)
            
```
